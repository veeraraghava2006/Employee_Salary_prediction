# Diffrenent Types of machine learning alogrithm

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler, OneHotEncoder

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

models = {
    "LogisticRegression": LogisticRegression(),
    "RandomForest": RandomForestClassifier(),
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(),
    "GradientBoosting": GradientBoostingClassifier(),
     "DecisionTree": DecisionTreeClassifier(),
    "neural_network": MLPClassifier(max_iter=200),
    "naive_bayes": GaussianNB()
}

results = {}

for name, model in models.items():
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        ('model', model)
    ])

    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results[name] = acc
    print(f"{name} Accuracy: {acc:.4f}")
    print(classification_report(y_test, y_pred))
```

# visualization

```python
import matplotlib.pyplot as plt
plt.bar(results.keys(), results.values(), color='skyblue')
plt.ylabel('Accuracy Score')
plt.title('Model Comparison')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

# Finding best machine learning alogrithm

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import joblib

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Define models
models = {
    "LogisticRegression": LogisticRegression(max_iter=1000),
    "RandomForest": RandomForestClassifier(),
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(),
    "GradientBoosting": GradientBoostingClassifier(),
     "DecisionTree": DecisionTreeClassifier(),
    "neural_network": MLPClassifier(max_iter=200),
    "naive_bayes": GaussianNB()
}

results = {}

# Train and evaluate
for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    acc = accuracy_score(y_test, preds)
    results[name] = acc
    print(f"{name}: {acc:.4f}")

# Get best model
best_model_name = max(results, key=results.get)
best_model = models[best_model_name]
print(f"\n Best model: {best_model_name} with accuracy {results[best_model_name]:.4f}")

# Save the best model
joblib.dump(best_model, "best_model.pkl")
print(" Saved best model as best_model.pkl")
```

```python


%%writefile app.py
import streamlit as st
import pandas as pd
import joblib


try:
    model = joblib.load("best_model.pkl")
    st.success("Model sucessfuly load")
except FileNotFoundError:
    st.error("Model file 'best_model.pkl' not found. Please train and save the model first.")
    st.stop()

st.set_page_config(page_title="Employee Salary Classification", page_icon="", layout="centered")

st.title(" Employee Salary Classification App")
st.markdown("Predict whether an employee earns >50K or 50K based on input features.")


st.sidebar.header("Input Employee Details")


age = st.sidebar.slider("Age", 17, 75, 30)

work_class_raw = st.sidebar.selectbox("Work Class", [
    "Private", "Self-emp-not-inc", "Local-gov", "Others", "State-gov", "Self-emp-inc", "Federal-gov"
])

educational_num = st.sidebar.slider("Education Number", 5, 16, 10)


occupation_raw = st.sidebar.selectbox("Occupation", [
    "Tech-support", "Craft-repair", "Other-service", "Sales",
    "Exec-managerial", "Prof-specialty", "Handlers-cleaners", "Machine-op-inspct",
    "Adm-clerical", "Farming-fishing", "Transport-moving", "Priv-house-serv",
    "Protective-serv", "Armed-Forces", "Others"
])


race_raw = st.sidebar.selectbox("Race", [
    "White", "Black", "Asian-Pac-Islander", "Amer-Indian-Eskimo", "Other"
])
gender_raw = st.sidebar.selectbox("Gender", ["Male", "Female"])
capital_gain = st.sidebar.number_input("Capital Gain", min_value=0, max_value=100000, value=0)
capital_loss = st.sidebar.number_input("Capital Loss", min_value=0, max_value=100000, value=0)
hours_per_week = st.sidebar.slider("Hours per week", 1, 99, 40)
native_country_raw = st.sidebar.selectbox("Native Country", [
    "United-States", "Cambodia", "England", "Puerto-Rico", "Germany", "Outlying-US(Guam-USVI-etc)",
    "India", "Japan", "Greece", "South", "China", "Cuba", "Iran", "Honduras", "Philippines",
    "Italy", "Poland", "Columbia", "Canada", "Thailand", "Ecuador", "Haiti", "Portugal",
    "Dominican-Republic", "El-Salvador", "France", "Guatemala", "Mexico", "Trinidad&Tobago",
    "Nicaragua", "Peru", "Vietnam", "Hungary", "Ireland", "Scotland", "Hong", "Holand-Netherlands"
])


input_data_raw = {
    'age': [age],
    'workclass': [work_class_raw],

    'educational-num': [educational_num],

    'occupation': [occupation_raw],

    'race': [race_raw],
    'gender': [gender_raw],
    'capital-gain': [capital_gain],
    'capital-loss': [capital_loss],
    'hours-per-week': [hours_per_week],
    'native-country': [native_country_raw]
}

input_df_raw = pd.DataFrame(input_data_raw)

st.write("###  Input Data (Raw)")
st.write(input_df_raw)


input_df_processed = input_df_raw.copy()

try:

    # Based on value counts and replacements:
    workclass_map = {"Private": 0, "Self-emp-not-inc": 1, "Local-gov": 2, "Others": 3, "State-gov": 4, "Self-emp-inc": 5, "Federal-gov": 6}
    marital_status_map = {"Never-married": 0, "Married-civ-spouse": 1, "Divorced": 2, "Married-spouse-absent": 3, "Separated": 4, "Married-AF-spouse": 5, "Widowed": 6}
    occupation_map = {
        "Tech-support": 0, "Craft-repair": 1, "Other-service": 2, "Sales": 3,
        "Exec-managerial": 4, "Prof-specialty": 5, "Handlers-cleaners": 6,
        "Machine-op-inspct": 7, "Adm-clerical": 8, "Farming-fishing": 9,
        "Transport-moving": 10, "Priv-house-serv": 11, "Protective-serv": 12,
        "Armed-Forces": 13, "Others": 14
    } # Order  depends on original data
    relationship_map = {"Not-in-family": 0, "Husband": 1, "Wife": 2, "Own-child": 3, "Unmarried": 4, "Other-relative": 5}
    race_map = {"White": 0, "Black": 1, "Asian-Pac-Islander": 2, "Amer-Indian-Eskimo": 3, "Other": 4}
    gender_map = {"Female": 0, "Male": 1}

    input_df_processed['workclass'] = input_df_processed['workclass'].map(workclass_map)
    input_df_processed['marital-status'] = input_df_processed['marital-status'].map(marital_status_map)
    input_df_processed['occupation'] = input_df_processed['occupation'].map(occupation_map)
    input_df_processed['relationship'] = input_df_processed['relationship'].map(relationship_map)
    input_df_processed['race'] = input_df_processed['race'].map(race_map)
    input_df_processed['gender'] = input_df_processed['gender'].map(gender_map)


    input_df_processed['native-country'] = 39

except Exception as e:
    st.error(f"Error during preprocessing (encoding): {e}")
    st.stop()


input_df_processed.fillna(-1, inplace=True)




expected_columns = [
    'age', 'workclass', 'fnlwgt', 'educational-num', 'marital-status',
    'occupation', 'relationship', 'race', 'gender', 'capital-gain',
    'capital-loss', 'hours-per-week', 'native-country'
]

try:


    input_df_final = pd.DataFrame(columns=expected_columns)


    for col in expected_columns:
        if col in input_df_processed.columns:
            input_df_final[col] = input_df_processed[col]
        else:

            st.warning(f"Missing column in input: {col}. Adding with value 0.")
            input_df_final[col] = 0 # Or use the mean/median from training data

    st.write("###  Processed Input Data (for Model)")
    st.write(input_df_final)

except Exception as e:
    st.error(f"Error during final DataFrame preparation: {e}")
    st.stop()


# Predict button
if st.button("Predict Salary Class"):
    try:
        # Make sure input_df_final has the correct format and order for the model
        prediction = model.predict(input_df_final)


        predicted_class = ">50K" if prediction[0] == 1 else "<=50K"
        st.success(f" Predicted Salary Class: **{predicted_class}**")
    except Exception as e:
        st.error(f"Error during prediction: {e}")

# Batch prediction
st.markdown("---")
st.markdown("####  Batch Prediction")
uploaded_file = st.file_uploader("Upload a CSV file for batch prediction", type="csv")

if uploaded_file is not None:
    try:
        batch_data_raw = pd.read_csv(uploaded_file)
        st.write("Uploaded data preview:", batch_data_raw.head())


        batch_data_processed = batch_data_raw.copy()


        batch_data_processed['workclass'] = batch_data_processed['workclass'].map(workclass_map)
        batch_data_processed['marital-status'] = batch_data_processed['marital-status'].map(marital_status_map)
        batch_data_processed['occupation'] = batch_data_processed['occupation'].map(occupation_map)
        batch_data_processed['relationship'] = batch_data_processed['relationship'].map(relationship_map)
        batch_data_processed['race'] = batch_data_processed['race'].map(race_map)
        batch_data_processed['gender'] = batch_data_processed['gender'].map(gender_map)


        batch_data_processed.fillna(-1, inplace=True)


        batch_data_final = pd.DataFrame(columns=expected_columns)
        for col in expected_columns:
            if col in batch_data_processed.columns:
                batch_data_final[col] = batch_data_processed[col]
            else:
                 st.warning(f"Missing column in batch input: {col}. Adding with value 0.")
                 batch_data_final[col] = 0 # Or use the mean/median from training data

        st.write("Processed batch data preview:", batch_data_final.head())

        batch_preds = model.predict(batch_data_final)

        # Map predictions back to original labels
        predicted_classes = [">50K" if pred == 1 else "<=50K" for pred in batch_preds]

        batch_data_raw['PredictedClass'] = predicted_classes # Add prediction to the original raw data

        st.write(" Predictions:")
        st.write(batch_data_raw.head())

        # Provide download link
        csv = batch_data_raw.to_csv(index=False).encode('utf-8')
        st.download_button(
            label="Download Predictions CSV",
            data=csv,
            file_name='predicted_classes.csv',
            mime='text/csv'
        )
    except Exception as e:
        st.error(f"Error during batch prediction: {e}")

```
